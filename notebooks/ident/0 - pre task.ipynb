```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sympy.plotting import plot

from scipy import integrate
```

```python
a1 = 2.; a2 = 3.
b1 = 4.; b2 = 4.
c1 = .5; c2 = 2.
k1 = 4.; k2 = 1.5
def systemA(N, t, i):
    return np.array([ a1*N[0]*(1 - b1*N[1] - N[0]/k1),
                      a2*N[1]*(1 - (i*0.025+9.2)*N[0] - N[1]/k2  )])
```

```python
print((k1-b1*k1*k2)/(1-b1*b2*k1*k2), (k2-b2*k1*k2)/(1-b1*b2*k1*k2))
```

```python
time = np.linspace(0,30,1000) 
initialsConditions = np.array([5, 20])   

plt.figure(figsize=(20,20))
for i in range(1,11):
    plt.subplot(5,5,i)
    
    onditions, infodict = integrate.odeint(systemA, initialsConditions, time, (i,) ,full_output=True)
    populationOfN1, populationOfN2 = onditions.T
    
    plt.plot(time, populationOfN2, 'b-', label='population Of N2')
    plt.plot(time, populationOfN1, 'r-', label='population Of N1')
    
    plt.title("k2={0}".format( (i*0.25 ) )   )

    plt.legend()
    plt.grid(True)
```

```python
time = np.linspace(0,60,1000) 
initialsConditions = np.array([5, 20])   

plt.figure(figsize=(20,20))
for i in range(1,11):
    plt.subplot(5,5,i)
    
    x,y = np.linspace(0,60,10),np.linspace(0,30,10)
    X,Y = np.meshgrid(x,y)

    U = a1*X*(1 - b1*Y - X/k1)
    V = a2*Y*(1 - b2*X - Y/(i*0.25)   )

    speed = np.sqrt(U*U + V*V)
    start = [[5,20]]

    plt.title("k2={0}".format(i*0.25)  )
    plt.streamplot(x,y, U, V, color=(.75,.90,.93))
    plt.streamplot(x,y, U, V, start_points=start, color="crimson", linewidth=2)
    plt.plot(0., k2, color='blue', marker='o', linestyle='dashed', linewidth=2, markersize=5)
    plt.plot(5, 20, color='blue', marker='o', linestyle='dashed', linewidth=2, markersize=5)

```
